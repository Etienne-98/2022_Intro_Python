#####**Task0: Data clean up. Remove all rows in the table if the rows contain missing values**

Hint: `pandas` has a easy function for data clean up. Check out [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html) for details.

```python
penguin_df = (penguins.dropna()) # drop the Not a number - values
print('show penguin_df:\n', penguin_df)
print('the size of my dataset is:\n', penguin_df.shape) # note there is no parentheses; tuple not callable
```

#####**Task1: Create an input dataset `X` with the culmen_length_mm and flipper_length_mm data columns**

Hints: 

(1) The shape of your input data should be `(334, 2)`

```python
# Create your input for model training here
# Input should contain penguin_df['culmen_length_mm'] and penguin_df['flipper_length_mm']
X = np.vstack((penguin_df['culmen_length_mm'], penguin_df['flipper_length_mm'])).T  # apparently T stands for 'transpose'
print('shape of X is:\n', 
      X.shape) # we want indeed to have a (334,2); without the T we'd end up having (2,334).
```

#####**Task2: Train a k-means clustering algorithm, perform elbow test and silhouette analysis**

Hints: 

(1) The documentation for KMeans clustering can be found [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html?highlight=kmeans#sklearn.cluster.KMeans)

(2) [Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html?highlight=silhouette+score#sklearn.metrics.silhouette_score) for silhouette score analysis in scikit-learn

```python
# Import KMeans fron scikit-learn
from sklearn.cluster import KMeans                    # usual dowmload procedure
# Import Silhouette score from scikit-learn
from sklearn.metrics import silhouette_score          # idem
```

```python
# Store the K-means inertia in an empty list
Sum_squared_d = []
K = range(1,10)
for i in K :                     # for i cluster in K (reminder K= [1, 2, 3, 4, 5, 6, 7, 8, 9])
 kmeans = KMeans(n_clusters = i) # implement algorithm
 kmeans.fit(X)                   # always do the fitting
 Sum_squared_d.append(kmeans.inertia_)
```

```python
print(list(range(1,10)))
```

```python
# To finish the elbow method analysis, plot the change in inertia when you change the number of clusters you used to train the k-means clustering algorithm.
plt.plot(K,
         Sum_squared_d,
         marker='s',c='k',lw=2)
plt.xlabel('Number of Clusters') 
plt.ylabel('Sum of Squared Distances / Inertia') 
plt.title('Elbow Method\nfor Optimal K')
plt.show()
```

```python
# Import silhouette_score for analysis
# from sklearn.metrics import silhouette_score  already imported
```

```python
# Perform silhouette analysis following instructions in the tutorial notebook
sil_average = []                    # create an empty list
for num_clusters in range(2,10):
  # initialise kmeans
  kmeans = KMeans(n_clusters=num_clusters)   # initialising
  kmeans.fit(X)                              # fitting
  cluster_labels = kmeans.labels_            # labelling
  
  # silhouette score
  sil_average.append(silhouette_score(X, cluster_labels))
```

```python
# Plot your silhouette analysis result here.
plt.plot(range(2,10),
         sil_average, 
         marker='s', c='b', lw=2)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette scores')
plt.title('Silhouette analysis\nfor optimal K')
```

##### **Task2: Perform KMeans clustering with 3 clusters**

```python
# Train a k-means clustering model
kmeans = KMeans(n_clusters=3)           # specify number of clusters
kmeans.fit(X)

# Use the model to label all data points
ykmeans = kmeans.predict(X)             # prediction y = ax+b
```

Now, our penguin dataset is labelled. So we can compare our clustering results with the truth to see how well our trained model performed.
